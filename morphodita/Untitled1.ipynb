```python
import sys
from corpy.morphodita import Tagger
from inspect import getmembers, isfunction
import json
import pandas as pd
import openpyxl
```

```python
tags = open('pdt_tags.json')
tags_data = json.load(tags)
text = open('text_file.txt','r').read()
```

```python
tagger = Tagger('slovak-morfflex-pdt-170914.tagger')
```

```python
testdf = pd.DataFrame(columns=['Slovo', 'Slovný druh', 'Rod', 'Číslo', 'Pád', 'Paradigma', 'Stupeň', 'Aglutinovanosť', 'Slovesná forma', 'Vid', 'Osoba', 'Negácia','Druh','Forma','Spojenie s pádom', 'Kondicionálnosť'])
```

```python
doctest = list(tagger.tag("Je mi zima. Bude snežiť."))
```

```python
doctest
```

```python
for word in doctest:
    row = {}
    row['Slovo'] = word.word
    for i,letter in enumerate(word.tag):
        if not letter == '-':
            if str(i) in tags_data:
                row[tags_data[str(i)][letter]['label']] = tags_data[str(i)][letter]['value']
    testdf = testdf.append(row, ignore_index = True)
```

```python
testdf.to_csv('testmorpho.csv', sep=';',encoding='utf-8-sig',index=False)
```
