```python
import sys
from corpy.morphodita import Tagger
from inspect import getmembers, isfunction
import json
import pandas as pd
import openpyxl
```

```python
tags = open('pdt_tags.json')
tags_data = json.load(tags)
text = open('text_file.txt','r').read()
```

```python
tagger = Tagger('slovak-morfflex-pdt-170914.tagger')
```

```python
from pprint import pprint
tokens = list(tagger.tag("Je mi zima. Bude snežiť."))
pprint(tokens)
```

```python
df = pd.DataFrame(columns=['Slovo', 'Slovný druh', 'Rod', 'Číslo', 'Pád', 'Paradigma', 'Stupeň', 'Aglutinovanosť', 'Slovesná forma', 'Vid', 'Osoba', 'Negácia','Druh','Forma','Spojenie s pádom', 'Kondicionálnosť'])
```

```python
doc = list(tagger.tag(text))
```

```python
doc
```

```python
for word in doc:
    row = {}
    row['Slovo'] = word.word
    for i,letter in enumerate(word.tag):
        try:
            if not letter == '-':
                if str(i) in tags_data:
                    row[tags_data[str(i)][letter]['label']] = tags_data[str(i)][letter]['value']
        except:
            print(word.tag, word.word)
    df = df.append(row, ignore_index = True)
            
```

```python
df.to_csv('morphodita_anotacia.csv', sep=';',encoding='utf-8-sig',index=False)
```
