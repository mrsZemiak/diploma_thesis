```python
import json
import pandas as pd
import openpyxl
```

```python
tags = open('tags-treetagger.json')
tags_data = json.load(tags)
text = open('rnntaggerout.txt','r').read()
```

```python
text
```

```python
df = pd.DataFrame(columns=['Slovo', 'Slovný druh', 'Rod', 'Číslo', 'Pád', 'Paradigma', 'Stupeň', 'Aglutinovanosť', 'Slovesná forma', 'Vid', 'Osoba', 'Negácia','Druh','Forma','Spojenie s pádom', 'Kondicionálnosť'])
```

```python
test = 'Kde\tPD\tkde\ntá\tPFfs1\ttá\nľudská\tAAfs1x\tľudský\nduša\tSSfs1\tduša\ndrieme\tVKesc+\tdriemať\n?\tZ\t?\n\nPrečo\tPD\tprečo\nto\tPFns4\tto\ntu\tPD\ttu\npíšem\tVKesa+\tpísať\n?\tZ\t?\n\n'
```

```python
sentences=text.split('\n')
sentences.pop()
```

```python
for sentence in sentences:
    word=sentence.split('\t')
    row={}
    if len(word) > 1:
        row['Slovo']=word[0]
        if word[1] == 'Z':
            row['Slovný druh'] = 'Interpunkcia'
        else:
             for i,letter in enumerate(word[1]):
                if i == 0:
                    row['Slovný druh'] = tags_data[letter]['slovakName']
                else: 
                    if bool(tags_data[word[1][0]]['tags']):
                        if letter in tags_data[word[1][0]]['tags']:
                            row[tags_data[word[1][0]]['tags'][letter]['label']] = tags_data[word[1][0]]['tags'][letter]['value']
                        else:
                            if "r" == word[1][i]:
                                row['Informácia'] = 'Príslovka'
                                break
                            if ":" == word[1][i]:
                                if ":r" == word[1][i:i+2]:
                                    row['Informácia'] = 'Vlastné meno'
                                    break

                                else:
                                    row['Informácia'] = 'Chybný zápis'
                                    break
        df = df.append(row, ignore_index = True)
            
```
