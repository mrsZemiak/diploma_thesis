```python
df2 = snkdf.iloc[0:1000, 0:2]
```

```python
df = stanzadf.iloc[0:1000, 0:2]
```

```python
final = pd.DataFrame(columns=['Slovo', 'Slovný druh', 'Slovo2', 'Slovný druh2'])
```

```python
k = 0
for i, slovo in df.iterrows():
    row = {}
    row['Slovo'] = slovo.Slovo
    row['Slovný druh'] = slovo['Slovný druh']
    if slovo.Slovo == df2['Slovo'][i+k]:
        row['Slovo2'] = df2.Slovo[i+k]
        row['Slovný druh2'] = df2['Slovný druh'][i+k]
    else:
        print(slovo.Slovo, df2['Slovo'][i+k])
        
    final = final.append(row, ignore_index = True)
```
